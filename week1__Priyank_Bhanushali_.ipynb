### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```

     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.

3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
# YOUR CODE HERE
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    left = A @ v
    right = lam * v

    print(f"\nEigenpair {i+1}")
    print("A @ v =", left)
    print("λ * v =", right)

    assert np.allclose(left, right)

# Plot
# YOUR CODE HERE
fig, ax = plt.subplots(figsize=(6,6))

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    ax.quiver(
        0, 0,
        lam * v[0],
        lam * v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.legend()

plt.grid(True)
plt.show()


print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

A matrix transformation can rotate, stretch, compress, or shear vectors in space. An eigenvector is a special vector whose direction remains unchanged after the transformation. The corresponding eigenvalue tells us how much the vector is stretched or compressed.

If the eigenvalue is greater than 1, the vector is stretched. If it is between 0 and 1, the vector is compressed. If it is negative, the vector is flipped in direction and scaled by the magnitude of the eigenvalue.

Therefore, eigenvectors represent the principal directions of a transformation, and eigenvalues represent the scaling factors along those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False) # YOUR CODE HERE

X_reconstructed =  U @ np.diag(S) @ Vt  # YOUR CODE HERE — reconstruct from U, S, Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

### Relationship Between SVD and PCA

Singular Value Decomposition (SVD) factorizes a matrix X into three matrices:

X = UΣVᵀ

where:

* U contains the left singular vectors.
* Σ contains the singular values.
* Vᵀ contains the right singular vectors.

Principal Component Analysis (PCA) uses these singular vectors to identify directions of maximum variance in the data. The principal components correspond to the columns of V (or rows of Vᵀ), while the singular values indicate the importance of each component.

Dimensionality reduction is achieved by keeping only the largest singular values and their corresponding singular vectors. This preserves most of the information while reducing the number of dimensions. The rank-1 approximation is the simplest example, where only the most important singular component is retained.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller
from scipy import stats
import numpy as np

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS Test against standard normal distribution
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts) # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE
    epsilon = 1e-10

    breakpoints = np.linspace(
        np.min(expected),
        np.max(expected),
        bins + 1
    )

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    expected_pct = expected_pct + epsilon
    actual_pct = actual_pct + epsilon

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
# YOUR CODE HERE
ax.hist(
    train_dist,
    bins=30,
    alpha=0.6,
    density=True,
    label='Training Distribution'
)

ax.hist(
    drift_dist,
    bins=30,
    alpha=0.6,
    density=True,
    label='Production Distribution'
)

ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**



#### Concept Drift

Concept drift occurs when the relationship between input features and the target variable changes over time. As a result, the model's learned patterns become outdated.

#### Covariate Drift

Covariate drift occurs when the distribution of input features changes while the relationship between features and the target remains unchanged.

#### Population Stability Index (PSI)

PSI is commonly interpreted as:

* PSI < 0.10 → No significant shift (Stable)
* 0.10 ≤ PSI < 0.25 → Moderate or Minor shift
* PSI ≥ 0.25 → Major shift (Investigate and potentially retrain)

#### Retraining Trigger

One common retraining trigger is when PSI exceeds 0.25 for critical features or when production model performance drops below an acceptable threshold.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam +
    p_free_given_ham * p_ham
)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free  # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    prior_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam +
        p_word_given_ham * prior_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([
    np.mean(
        np.random.choice(
            population,
            size=sample_size,
            replace=True
        )
    )
    for _ in range(n_samples)
])
 # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)  # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(
    sample_means,
    bins=40,
    density=True,
    alpha=0.7,
    label='Sample Means'
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    500
)

ax.plot(
    x,
    norm.pdf(x, pop_mean, clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**
### Central Limit Theorem (CLT)

The Central Limit Theorem states that regardless of the shape of the original population distribution, the distribution of sample means approaches a normal distribution as the sample size becomes sufficiently large.

In this experiment:

* The population follows an exponential distribution, which is highly skewed.
* We repeatedly draw samples of size 30.
* The distribution of the resulting sample means becomes approximately normal.

The theoretical mean of the sampling distribution equals the population mean, and the theoretical standard deviation equals:

σₓ̄ = σ / √n

This explains why many statistical methods assume normality when working with sample means.
